```python
import requests
from bs4 import BeautifulSoup

# Inisialisasi list istilah
all_terms = []

# Loop semua halaman glossary
for page in range(3):  # 0, 1, 2
    url = f"https://birds.iitk.ac.in/glossary?page={page}"
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    # Ambil semua elemen <strong> dalam tabel
    for strong in soup.select("table strong"):
        term = strong.get_text(strip=True).replace('\xa0', '')  # Hilangkan spasi non-breaking
        if term:
            all_terms.append(term)

# Hilangkan duplikat dan urutkan
unique_terms = sorted(set(all_terms))

# Tampilkan hasil
for term in unique_terms:
    print(term)

# Simpan ke file opsional
with open("glossary_terms.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(unique_terms))
```

```python
import requests
from bs4 import BeautifulSoup

# URL glossary
url = "https://academy.allaboutbirds.org/bird-academys-a-to-z-glossary-of-bird-terms/"

# Header untuk menghindari 403 Forbidden
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# Ambil halaman
response = requests.get(url, headers=headers)
response.raise_for_status()

# Parsing HTML
soup = BeautifulSoup(response.text, "html.parser")

# Ambil semua <dt> sebagai istilah
terms = [dt.get_text(strip=True) for dt in soup.find_all("dt")]

# Simpan ke file
with open("bird_glossary_terms.txt", "w", encoding="utf-8") as f:
    f.write("\n".join(terms))

print("✅ Istilah berhasil disimpan ke 'bird_glossary_terms.txt'")
```
