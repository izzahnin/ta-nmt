```python
import os
import json
import re

# Folder tempat file JSON disimpan
FOLDER_PATH = "paralel_disempurnakan"

# Kata kunci yang ingin dicari (tambahkan atau ubah sesuai kebutuhan)
keywords = ["eyering"]  # ← ubah di sini

# Buat pola regex untuk tiap keyword
patterns = [re.compile(rf'{kw}\w*', re.IGNORECASE) for kw in keywords]

# Fungsi untuk mencari kata di caption
def search_in_caption(text, patterns):
    return any(pattern.search(text) for pattern in patterns)

# Proses setiap file JSON
for filename in sorted(os.listdir(FOLDER_PATH)):
    if filename.endswith(".json"):
        filepath = os.path.join(FOLDER_PATH, filename)
        try:
            with open(filepath, "r", encoding="utf-8") as f:
                content = json.load(f)
        except Exception as e:
            print(f"[ERROR reading {filename}]: {e}")
            continue

        # Deteksi struktur
        if isinstance(content, dict) and "dataset" in content:
            entries = content["dataset"]
        elif isinstance(content, list):
            entries = content
        else:
            print(f"[WARNING] Struktur tidak dikenali di {filename}")
            continue

        for item in entries:
            for cap in item.get("captions", []):
                eng = cap.get("english", "")
                indo = cap.get("indo", "")
                if search_in_caption(eng, patterns) or search_in_caption(indo, patterns):
                    print(f"[MATCH in {filename}]")
                    print("ENG :", eng)
                    print("IND :", indo)
                    print("-" * 80)
```
