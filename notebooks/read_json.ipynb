```python
import json
from collections import defaultdict
import os

# Load data JSON dari file
with open("paralel_cub_200_2011_captions_disempurnakan_updated.json", "r", encoding="utf-8") as f:
    data = json.load(f)["dataset"]  # asumsi struktur awal adalah {"dataset": [...]}

# Buat dictionary berdasarkan class_id
class_map = defaultdict(list)
for item in data:
    class_id = item["class_id"]
    class_map[class_id].append(item)

# Pastikan folder output ada
os.makedirs("paralel_disempurnakan", exist_ok=True)

# Simpan tiap class_id ke file terpisah
for class_id in range(200):  # 0 hingga 199
    entries = class_map.get(class_id, [])
    if entries:
        with open(f"paralel_disempurnakan/class_{class_id}.json", "w", encoding="utf-8") as f:
            json.dump({"dataset": entries}, f, ensure_ascii=False, indent=4)
```

```python
import re

def clean_text(text):
    # Hilangkan karakter selain huruf, angka, dan tanda baca dasar
    text = re.sub(r"[^a-zA-Z0-9.,;:'\"()\-\s]", '', text)
    # Hilangkan spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def remove_duplicates(data):
    seen = set()
    cleaned = []
    for item in data:
        pair = (item['english'], item['indo'])
        if pair not in seen:
            seen.add(pair)
            cleaned.append(item)
    return cleaned
```
